Some examples from notes: square root inverse, $y=1/\sqrt(x+1)$

Domain for this function is $(-1,\infty ] $

To avoid overflow in drawing the graph, we specify `endpoint=False` and write the interval as `5, -1` so that the endpoint is applied to `-1`.

```python
x=np.linspace(5,-1, endpoint=False)
y=1/(np.sqrt(x +1) )
plt.plot(x, y)
plt.title("Function example")
```

Sketch a function 

```python
x=np.linspace(-2,3,6)
y=1+x**2
x1=np.linspace(-2,3)
y1=1+x1**2
plt.plot(x1,y1)
plt.plot(x,y,'o')
plt.title("Sketched Graph $y=1+x^2$")
```

# Now some graphs
Sketch a directed graph:

$ N=(1,2,3,4,5) $

$ E=\{(1,2) ,(1,3),(3,3),(4,2),(5,4),(4,3)\} $



Set up node options first

```python
options = {
    "font_size": 16,
    "node_size": 600,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 2,
    "width": 2,

}
```

```python
G=nx.DiGraph()
G.add_edges_from([(1,2),(1,3),(3,3),(4,2),(5,4),(4,3)])
nx.draw_planar (G, with_labels=True, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()

# From https://networkx.org/documentation/stable/auto_examples/basic/plot_simple_graph.html
```
